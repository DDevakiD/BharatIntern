```python
from sklearn.preprocessing import LabelEncoder
categ = ["date","price","bedrooms","bathrooms","sqft_living","sqft_lot","floors","waterfront","view","condition","sqft_above","sqft_basement","yr_built","yr_renovated","street","city","statezip","country"]

# Encode Categorical Columns
le = LabelEncoder()
Price[categ] = Price[categ].apply(le.fit_transform)
```

## Splitting Train and Test Data

```python
X =Price.drop(['price'],axis=1)
X
```

```python
Y =Price['price']
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=21)
```

```python
len(X_train)
```

```python
len(X_test)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_predict = model.predict(X_train)
X_predict
```

```python
Y_predict = model.predict(X_test)
Y_predict
```

## Logistic Regression

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
lr_classifier = LogisticRegression()
lr_classifier.fit(X_train, Y_train)
```

```python
train_pred = lr_classifier.predict(X_train)
test_pred  = lr_classifier.predict(X_test)
```

```python
accuracy_score(Y_train, train_pred)
```

```python
accuracy_score(Y_test, test_pred)
```

```python
confusion_matrix(Y_train, train_pred)
```

```python
confusion_matrix(Y_test, test_pred)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_classifier = RandomForestClassifier(n_estimators = 100, max_depth = 12, random_state = 21)
rf_classifier.fit(X_train, Y_train)
```

## Accuracy

```python
train_pred = rf_classifier.predict(X_train)
```

```python
test_pred  = rf_classifier.predict(X_test)
```

```python
accuracy_score(Y_train, train_pred)
```

```python
accuracy_score(Y_test, test_pred)
```
