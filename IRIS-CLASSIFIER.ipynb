## Data Visualization

```python
Iris.hist(bins=50, figsize=(20,20))
```

```python
from sklearn.preprocessing import LabelEncoder
categ = ["Id","SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]

# Encode Categorical Columns
le = LabelEncoder()
Iris[categ] = Iris[categ].apply(le.fit_transform)
```

```python
corr = Iris.corr()
plt.figure(figsize=(12,7))
sb.heatmap(corr,cmap='coolwarm',annot=True)
```

## Splitting Train and Test Data

```python
X = Iris.drop(['Species'],axis=1)
```

```python
X
```

```python
Y = Iris['Species']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=21)
```

```python
len(X_train)
```

```python
len(X_test)
```

## Logistic Regression

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
lr_classifier = LogisticRegression()
lr_classifier.fit(X_train, Y_train)
```

```python
train_pred = lr_classifier.predict(X_train)
test_pred  = lr_classifier.predict(X_test)
```

```python
confusion_matrix(Y_train, train_pred)
```

```python
confusion_matrix(Y_test, test_pred)
```

```python
accuracy_score(Y_train, train_pred)
```

```python
accuracy_score(Y_test, test_pred)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_classifier = RandomForestClassifier(n_estimators = 100, max_depth = 12, random_state = 21)
rf_classifier.fit(X_train, Y_train)
```

## Accuracy

```python
train_pred = rf_classifier.predict(X_train)
```

```python
test_pred  = rf_classifier.predict(X_test)
```

```python
accuracy_score(Y_train, train_pred)
```

```python
accuracy_score(Y_test, test_pred)
```
